```python
import random
from random import choices
import pandas as pd
import numpy as np
from pathlib import Path
import datetime
```

```python
# Load data
airbnb_data = pd.read_csv(Path("../raw_data/Airbnb_Texas_Rentals.csv"))
airbnb_df = pd.DataFrame(airbnb_data)
airbnb_df
```

```python
airbnb_df.reset_index(drop=True, inplace=True)
```

```python
airbnb_df= airbnb_df.drop(columns=["description", "url", "Unnamed: 0", "title"])
airbnb_df
```

```python
airbnb_df = pd.get_dummies(airbnb_df, columns=['city' ])
airbnb_df
```

```python
airbnb_df['date']= pd.to_datetime(airbnb_df['date_of_listing'])
airbnb_df.drop(columns=["date_of_listing"], inplace=True)
airbnb_df
```

```python
airbnb_df.set_index('date', inplace=True)
airbnb_df
```

```python
airbnb_df.sort_index(ascending=True, inplace=True)
airbnb_df
```

```python
# Saving the encoded dataset
file_path = Path("../../data/airbnb_TX_encoded.csv")
airbnb_df.to_csv(file_path, index=True)
```
